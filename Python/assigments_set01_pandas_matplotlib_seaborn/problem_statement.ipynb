```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
# 1. Load the frozen_confections_data.csv file into a Pandas DataFrame
frozen_data=pd.read_csv('frozen_confections_data.csv')
# 2. Display the first 5 rows of the DataFrame
print("First 5 rows of the DataFrame:")
print(frozen_data.head())
print("="*20)
# 3.Get a concise summary of the DataFrame, including the data types of each column and the number of non-null values
print("Summary of the data:")
print(frozen_data.info())
print("="*20)
# 4. Display the column and their datatypes
print("Data types of the DataFrame:")
print(frozen_data.dtypes)
print("="*20)
# 5.Check the dimensions of the DataFrame
print("No.of rows and columns of the DataFrame:")
print(frozen_data.shape)
      
```

## Part 2: Data Cleaning and Manipulation (Pandas)

**Objective:** Clean the data and prepare it for analysis.

6.  Calculate the number of missing values in each column.
    * **Step:** Use `.isnull()` followed by `.sum()`.
7.  Check if there are any duplicate rows in the DataFrame.
    * **Step:** Use `.duplicated()` and `.sum()`.
8.  Rename the `SuccessfulOrder` column to `OrderSuccess`.
    * **Step:** Use the `.rename()` method with the `columns` parameter.
9.  Create a new column named `ItemsCount` which is the sum of `ToppingsCount` and `ScoopsCount`.
    * **Step:** Add the two columns together.

    * **Step:** Use the `.groupby()` method on `PaymentMethod` and then apply `.mean()` to `TotalCost`.

```python
# 6. Calculate the number of missing vlaues in each column
print("sum of the number of missing values in each column")
print(frozen_data.isnull().sum())
print("="*20)
# 7. Check if there are any duplicate rows in the DataFrame
print("duplicate values in rows:")
print(frozen_data.duplicated().sum())
print("="*20)
# 8. Reaname the SuccessfulOrder column to OrderSuccess
print("DataFrame with renamed column:")
frozen_df=frozen_data.rename(columns={'SuccessfulOrder': 'OrderSuccess'})
print(frozen_df)
print("="*20)
# 9. Create a new column named ItemsCount which is the sum of ToppoinsCount and ScoopsCount.
print("Creating a new column named ItemsCount:")
frozen_df['ItemsCount']=frozen_df['ToppingsCount']+frozen_df['ScoopsCount']
print(frozen_df.head())
print("="*20)
# 10. Calculate the average Totalcost for each PaymentMethod
print("Average Total cost for each Payment Method")
frozen_set=frozen_df.groupby('PaymentMethod')['TotalCost'].mean()
print(frozen_set)
```

## Part 3: Basic Exploratory Data Analysis (EDA) and NumPy

**Objective:** Use descriptive statistics and NumPy to gain initial insights.

11. Calculate the mean, median, and standard deviation of `CustomerAge`.
    * **Step:** Use `.mean()`, `.median()`, and `.std()` on the `CustomerAge` column.
12. Find the most common `PaymentMethod` and `FlavorCategory`.
    * **Step:** Use the `.mode()` method.
13. Count the number of successful and unsuccessful orders.
    * **Step:** Use the `.value_counts()` method on the `OrderSuccess` column.
14. Use NumPy to calculate the percentage of successful orders.
    * **Step:** Calculate the sum of successful orders and divide by the total number of records.
15. Find the maximum `TotalCost` and the minimum `TotalCost` from the dataset.
    * **Step:** Use `.max()` and `.min()` on the `TotalCost` column.

---

```python
# 11. Calculate the mean ,median and standard deviation of CustomerAge
print('Mean of the Customer Age:')
mean_age=np.mean(frozen_df['CustomerAge'])
print(mean_age)
print("*" * 35)
print("Median of the  Customer Age:")
median_age=np.median(frozen_df['CustomerAge'])
print(median_age)
print("*" * 35)
print("Standard Deviation of the Customer Age")
sd_age=np.std(frozen_df['CustomerAge'])
print(sd_age)
print("*" * 35)
#12.Find the most common PaymentMethod and FlavorCategory
most_common_payment=frozen_df['PaymentMethod'].mode()# module 'numpy' has no attribute 'mode,we can use it in pandas
print(most_common_payment)
most_flavorcategory=frozen_df['FlavorCategory'].mode()
print(most_flavorcategory)
print("*" * 35)
#13.Count the number of successful and unsuccessful orders
print("Number of successful and unsuccessful orders")
order_counts=frozen_df['OrderSuccess'].value_counts()
print(order_counts)
print("*" * 35)
# 14.Use NumPy to calculate the percentage of successful orders
successful_orders=frozen_df['OrderSuccess'].sum()
total_orders=len(frozen_df['OrderSuccess'])
success_percentage=(successful_orders/total_orders)*100
print("Percentage of successful orders:")
print(success_percentage,"%")
print("*" * 35)
#15.Find the maximum TotalCost and the minimum TotalCost from the dataset
print("Maximum Total Cost:")
max_total_cost=np.max(frozen_df['TotalCost'])
print(max_total_cost)
print("Minimum Total Cost:")
min_total_cost=np.min(frozen_df['TotalCost'])
print(min_total_cost)
```

    * **Step:** Use `plt.scatter()` or `sns.scatterplot()`.

    * **Step:** Group the data by `PaymentMethod`, calculate the mean `TotalCost`, and then use `sns.barplot()`.

    * **Step:** Calculate the correlation matrix using `.corr()` and then use `sns.heatmap()`.

```python
# 16. Create a histogram of the CustomerAge distribution
plt.figure(figsize=(8,6))
plt.hist(frozen_df['CustomerAge'],bins=10,edgecolor='black',alpha=0.7)
plt.title('Distribution of Customer Age')
plt.xlabel('Age')
plt.ylabel('No.of Customers')
plt.show()
```

```python
# 17.Visualize the count of orders by PaymentMethod using a bar plot
plt.figure(figsize=(10,6))
sns.countplot(x='PaymentMethod',data=frozen_df)
plt.title('Count of orders by PaymentMethod')
plt.xlabel('PaymentMethod')
plt.ylabel('Count of orders')
plt.tight_layout()
plt.show()
```

```python
# 18. Create a countplot to show the number of successful orders (OrderSuccess) for each FlavorCategory
plt.figure(figsize=(10,6))
sns.countplot(x='FlavorCategory',hue='OrderSuccess',data=frozen_df,palette='viridis')
plt.title('successful orders (OrderSuccess) for each FlavorCategory')
plt.xlabel('FlavorCategory')
plt.ylabel('OrderSuccess')
plt.tight_layout()
plt.show()
```

```python
# 19.Create a box plot to show the distribution of TotalCost across different FlavorCategorys
plt.figure(figsize=(10,6))
sns.boxplot(x='FlavorCategory',y='TotalCost',data=frozen_df)
plt.title('Distribution of Total Cost across Different Flavor Categories')
plt.xlabel('Flavor Category')
plt.ylabel('Total Cost')
plt.tight_layout()
plt.show()
```

```python
#20. Create a violin plot to show the distribution of CustomerAge for successful versus unsuccessful orders
plt.figure(figsize=(10, 6))
sns.violinplot(x='OrderSuccess', y='CustomerAge',data=frozen_df,color='green') #In given data we have only one value i.e 0
plt.title('Distribution of CustomerAge for successful versus unsuccessful orders', fontsize=16)
plt.xlabel('OrderSuccess')
plt.ylabel('CustomerAge')
plt.tight_layout()
plt.show()
```

```python
#21. Generate a scatter plot showing the relationship between ToppingsCount and TotalCost
plt.figure(figsize=(8,6))
plt.scatter(frozen_df['ToppingsCount'],frozen_df['TotalCost'],color='Green',alpha=0.8)
plt.title('Relationship between ToppingCount and TotalCost')
plt.xlabel('ToppingCount')
plt.ylabel('TotalCost')
plt.show()

          
```

```python
# 22.Create a bar chart showing the average TotalCost per PaymentMethod
plt.figure(figsize=(10,6))
average_cost=frozen_df.groupby('PaymentMethod')['TotalCost'].mean()
sns.barplot(data=average_cost,palette=['green','blue','red'])
plt.title('Average Total Cost per Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Total Cost')
plt.tight_layout()
plt.show()
```

```python
#23. Visualize the correlation matrix of the numerical columns using a heatmap
numerical_columns=['FlavorCategory','CustomerAge','ToppingsCount','ScoopsCount','TotalCost','ItemsCount']
correlation_matrix=frozen_df[numerical_columns].corr()
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix,cmap='viridis',linewidth=(0.5))
plt.title('Correlation Matrix of Numerical Columns')
plt.show()
```

```python
#24.Create a pie chart to show the proportion of each FlavorCategory
plt.figure(figsize=(8,8))
flavor_counts=frozen_df['FlavorCategory'].value_counts()
plt.pie(flavor_counts,labels=flavor_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Counts of FlavorCategory')
plt.axis('equal')
plt.show()

```

```python
#25.Use a pair plot to visualize the relationships between all numerical features
sns.pairplot(frozen_df, hue='FlavorCategory', vars=['CustomerAge','ToppingsCount','ScoopsCount','TotalCost','ItemsCount'],palette="pastel")
plt.suptitle('Relationships between all numerical features', y=1.02, fontsize=16)
plt.show()
```
